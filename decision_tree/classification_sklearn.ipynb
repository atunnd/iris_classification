```python
import numpy as np
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt

# Defining a simple dataset
attribute_names =  ['age', 'income','student', 'credit_rate']
class_name = 'default'
data1 ={
    'age' : ['youth', 'youth', 'middle_age', 'senior', 'senior', 'senior','middle_age', 'youth', 'youth', 'senior', 'youth', 'middle_age','middle_age', 'senior'],
    'income' : ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium','low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student' : ['no','no','no','no','yes','yes','yes','no','yes','yes','yes','no','yes','no'],
    'credit_rate' : ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair','excellent', 'excellent', 'fair', 'excellent'],
    'default' : ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes','yes', 'yes', 'yes', 'no']
}
df = pd.DataFrame (data1, columns=data1.keys())
print(df)
```

```python
classifier = tree.DecisionTreeClassifier(criterion="gini")
```

```python
df.columns[:-1]
```

```python
one_hot_data = pd.get_dummies(df[['age', 'income', 'student', 'credit_rate']])

# Prepare the input and output for the classifier
X = one_hot_data.values
y = df['default'].values

# Fit the classifier
classifier.fit(X, y)

# Get the correct feature names after one-hot encoding
feature_names = one_hot_data.columns

# Plot the decision tree
fig, ax = plt.subplots(figsize=(20, 20))
tree.plot_tree(classifier, ax=ax, feature_names=feature_names, filled=True)

plt.show()
```
